```python
import pandas as pd
```

```python
pops = pd.read_csv('populations.csv')
```

```python
pops.describe()
```

```python
us_pop = pops[pops['NAME'] == 'United States'].values[0][5]
us_pop
```

```python
drop = ['Northeast Region', 'Midwest Region', 'West Region', 'South Region', 'Puerto Rico', 'United States']
pops = pops[~pops['NAME'].isin(drop)]
```

```python
pops = pops.sort_values('CENSUS2010POP')
pops
```

```python
def population(state, pops=pops):
    if (state == 'DC'):
        state = 'District of Columbia'
    return pops[pops['NAME'] == state].values[0][5]
```

```python
list(pops.sum())[5], us_pop
```

```python
votes_remaining_rounded = 538
votes_remaining_prop = 538
for index, row in pops.iterrows():
    votes = population(row['NAME']) / us_pop * 538
    if votes < 1:
        votes = 1
    votes_prop = min(votes_remaining_prop, votes)
    votes_round = round(min(votes_remaining_rounded, votes))
    
    votes_remaining_rounded -= votes_round
    votes_remaining_prop -= votes_prop
    pops.loc[index, 'prop_votes'] = votes_prop
    pops.loc[index, 'rounded_votes'] = votes_round

pops
```

```python
pops['rounded_votes'].sum()
```

```python
# give missouri an extra vote because they're lose the most votes (.43 lost)
pops.at[30, 'rounded_votes'] = 11
pops
```

```python
trump = 0
biden = 0
blue = ['California', 'New York', 'Illinois', 'Pennsylvania', 'Michigan', 'Georgia', 'Virginia', 'Washington', 'Massachusetts', 'Arizona', 'Maryland', 'Wisconsin', 'Minnesota', 'Colorado', 'Oregon', 'Connecticut', 'Nevada', 'New Mexico', 'Hawaii', 'New Hampshire', 'Rhode Island', 'Delaware', 'Vermont', 'District of Columbia']

r_states = []
b_states = []

no_winner = True
for index, row in reversed(list(pops.iterrows())):
    name = row['NAME']
    if name == 'Maine':
        trump += 1
        biden += 1
        r_states.append(name)
        b_states.append(name)
    elif name == 'Nebraska':
        trump += 2
        biden += 1
        r_states.append(name)
    else:
        if name in blue:
            biden += row['rounded_votes']
            b_states.append(name)
        else:
            trump += row['rounded_votes']
            r_states.append(name)
    if no_winner:
        if trump > 270:
            print("Trump won with ", ', '.join(r_states))
            no_winner = False
        elif biden > 270:
            print('Biden won with ', ', '.join(b_states))
            no_winner = False
            

print('Trump: ', trump)
print('Biden: ', biden)
```
